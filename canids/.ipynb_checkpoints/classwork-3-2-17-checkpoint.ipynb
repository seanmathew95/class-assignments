# occurences of canid species through time

```python
from collections import defaultdict
species_ranges = defaultdict(list)
canids = open("formatted_canids.csv")
canids.readline()
canid_dict= {}
for line in canids:
    species = line.split('","')[5]
    min_ma = float(line.split('","')[11])
    max_ma = float(line.split('","')[10])
    avg_ma = (max_ma + min_ma )/ 2
    
    species_ranges[species].append(avg_ma)
#new_canids = open("new_candis.csv", "w")
```

```python
species_ranges.values()
```

```python
#get the oldest and youngest for each species
output = open("canid_ranges.csv", "w")
for key in species_ranges.keys():
    ages = species_ranges[key]
    minage = min(ages)
    maxage = max(ages)
    genus = key.split(" ")[0]
    print(genus, key, minage, maxage)
    output.write(genus + "," + key + "," + str(minage) + "," + str(maxage) + "\n")
```
