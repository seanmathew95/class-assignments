```python
def makeBirdDict(filename):
    #opens a file and extracts species and family data and populates a dictionary
    birdDict ={}
    fobj = open(filename, "r", encoding = "ISO-8859-15") #open the file
    all_records = fobj.readlines()[2:] #read the file
    for line in all_records:
        species = line.split(",")[0]
        families = line.split(",")[2]
        birdDict[species] = families
    return birdDict
new_bird_dict =makeBirdDict("/home/eeb177-student/Desktop/eeb-177/class-assignments/nature.csv")
species =list(new_bird_dict.keys())
species.sort()
species
species[0]
print("Family of " + species[0] + " is " + new_bird_dict[species[0]])
```
